```python
import collections

# 64밀리초에 실행됨. 
# 거의 5배 가까이 빠른 속도를 확인 할 수 있음.
# pop(0)이 O(n)인 데 반해, deque의 popleft()는 O(1)이기 때문이며, 각각 n번씩 반복하면, 리스트 구현은O(n**2), 
# 데크 구현은 O(n)으로 성능 차이가 큼


def isPalindrome(s: str) -> bool:
# 자료형 데크로 선언
    strs: Deque = collections.deque()

    strs = []
    for char in s:
        if char.isalnum(): # 영문자, 숫자 여부를 판별하는 함수
            strs.append(char.lower())

        while len(strs) > 1:
            if strs.pop(0) != strs.pop(): # 리스트 함수 pop()은 인덱스를 지정 할 수 있음
                return False
        return True

s = 'aaaa'
isPalindrome(s)
```

### 스라이싱 사용

```python
import re

def isPalindrome(s:str) -> bool:
    s = s.lower()
    # 정규식을 이용 불필요한 문자 필터링
    s = re.sub('[^a-z0-9]', '', s)
    return s == s[::-1] # 슬라이싱
```
